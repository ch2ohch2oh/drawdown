```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

dat = yf.Ticker("MSFT")
```

```python
def get_data(ticker, period='5y'):
    data = yf.Ticker(ticker).history(period=period)
    data['RollingMax'] = data['Close'].rolling(window=20).max()
    data['Drawdown'] = data['Close']/data['RollingMax'] - 1
    data['1MTrailingReturn'] = data['Close']/data['Close'].shift(periods=20) - 1
    data['1YTrailingReturn'] = data['Close']/data['Close'].shift(periods=252) - 1
    return data

def add_to_fig(fig, data, ticker):
    fig.add_trace(go.Scatter(x=data.index, y=data['1MTrailingReturn'], name=ticker), row=1, col=1)
    fig.add_trace(go.Scatter(x=data.index, y=data['1YTrailingReturn'], name=ticker), row=2, col=1)
    fig.add_trace(go.Scatter(x=data.index, y=data['Drawdown'], name=ticker), row=3, col=1)

fig = make_subplots(rows=3, cols=1)

tickers = ['MSFT', 'AAPL', 'GOOG', 'QQQ']
data = {}
for ticker in tickers:
    data[ticker] = get_data(ticker, period='10y')
    add_to_fig(fig, data[ticker], ticker)
fig.show()
```
